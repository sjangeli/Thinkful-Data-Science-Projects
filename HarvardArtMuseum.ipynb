```
# import the requests library
import requests

# function for processing the data
def process_data(raw_names):
  # get the properties we want by iterating through the host dictionary
  return [{'displayname':name['displayname'], 
             'birthplace':name['birthplace']
           } for name in raw_names]

# declare an empty list to store all the British artists
brits = []

# declare some starter variables that will keep track of the size of the queries and what page we're on
page = 1
size = 100

# initial call
url = 'https://api.harvardartmuseums.org/person'
query = {'apikey':'406f36e0-19ca-11ea-a5d2-53c047326203','q':'culture:British', 'fields':'displayname,birthplace', 'size':size, 'page':page}
response = requests.get(url, params=query)
print('Querying {}'.format(response.url))

if(response.ok):
  # get the json data into a dictionary
  data = response.json()
  # see how many records there are in total, just for fun
  total = data['info']['totalrecords']
  # see how many pages there are in total, this is what I'll use to iterate through all the records
  num_pages = data['info']['pages']
  print('There is a total of {} results to fetch from {} pages'.format(total,num_pages))
  # put the current page of data into our storage list
  brits = process_data(data['records'])
  print('{} results processsed so far'.format(len(brits)))

# increment the page number once
page += 1

while page < num_pages+1:
  # do everything we did above again, but each time increment the page number once (this could replace the code above)
  query = {'apikey':'406f36e0-19ca-11ea-a5d2-53c047326203','q':'culture:British', 'fields':'displayname,birthplace', 'size':size, 'page':page}
  response = requests.get(url, params=query)
  print('Querying {}'.format(response.url))
  if(response.ok):
    brits = brits + process_data(data['records'])
    print('{} results processsed so far'.format(len(brits)))
    page += 1

print('{} results returned'.format(len(brits)))
print(brits)
```
