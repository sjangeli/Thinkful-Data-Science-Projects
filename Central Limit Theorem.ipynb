```
# For two of the following distributions:

# Normal
# Binomial
# Gamma
# Poisson

# Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.

# Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import normal
from numpy import mean
from scipy import stats
from numpy.random import seed
%matplotlib inline
```

```
# First Distribution: Normal - Plotting the average of the mean as trials -> 1,000

results = []
for num_trials in range(1,1000):
    trials = np.random.normal(5, 1, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()
```

```
# First Distribution: Normal - Plotting the means as a histogram

means = [mean(np.random.normal(5,1,100)) for i in range(1000)]
plt.hist(means)
plt.show()
```

```
results = []
for num_trials in range(1,1000):
    trials = np.random.gamma(5, 1, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})
print(df)


df.plot()
```

```
means = [mean(np.random.gamma(5,1,100)) for i in range(1000)]
plt.hist(means)
plt.show()
```
