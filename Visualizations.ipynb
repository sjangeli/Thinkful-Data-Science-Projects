```
# Loading the data...

# Code to read csv file into Colaboratory:
!pip install -U -q PyDrive
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
import pandas as pd

# Authenticate and create the PyDrive client.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

houseprices = drive.CreateFile({'id':'19ZAsw9E2MZdD60H8BlsIldd0afOS159u'}) 
houseprices.GetContentFile('houseprices.csv')  
houseprices = pd.read_csv('houseprices.csv')

# All the other imports

!pip install seaborn==0.9.0
import seaborn as sns 
sns.set()
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```
# 1. Illustrate the distribution of saleprice.

plt.figure(figsize = (15,10))
sns.distplot(houseprices['saleprice'], bins = 50)
plt.show()
```

```
# 2. Illustrate the distribution of the lotarea.

plt.figure(figsize = (20,10))
sns.distplot(houseprices['lotarea'], bins=100 ,color='red', kde=False)
plt.show()
```

```
# 3. Use a boxplot to illustrate how overallcond affects the sale price.

plt.figure(figsize = (11,5))
sns.boxplot(y="saleprice", x = "overallcond", data=houseprices)
plt.show()
```

```
# 4. On the same boxplot, highlight the difference between having central air conditioning or not. 

# You can see that if you have central air, you can generally get more for the house. 

plt.figure(figsize = (15,5))
sns.boxplot(y="saleprice", hue="centralair", x = "overallcond", data=houseprices)
plt.show()
```

```
# 5. Is there a correlation between lotarea and the saleprice? 

# Super weak correlation. 

plt.figure(figsize = (11,5))
sns.scatterplot(x="saleprice", y="lotarea", data=houseprices);
plt.show()
```
